<h2><a name="pengenalan-plotnine"></a>Chapter 1: Pengenalan Plotnine</h2>

Package plotnine ini berdasarkan pada implementasi package ggplot2 yang terdapat di bahasa R, dan menggunakan pendekatan konstruksi bernama grammar of graphics yang dicetuskan oleh Leland Wilkinson. Dengan pendekatan ini, kita dapat membuat visualisasi data baik yang simpel maupun kompleks dengan mudah dan menarik.

Plotnine sendiri dibuat di atas package matplotlib , yaitu salah satu package di Python yang umum digunakan untuk visualisasi data. Perbedaan antara plotnine dengan matplotlib adalah plotnine telah menyediakan berbagai fungsi yang mempermudah kita untuk membuat visualisasi data.

Package plotnine juga dapat digunakan dengan mudah bersamaan dengan pandas , package Python yang
biasanya digunakan untuk memanipulasi data.

<h3><a name="import-package"></a>Import Package</h3>

Pertama-tama, kita perlu mengimpor package plotnine terlebih dahulu.

Kita masih membutuhkan package pandas untuk membaca data serta melakukan manipulasi data sederhana sebelum kita membuat visualisasi data yang kita perlukan. Oleh karena itu kita perlu mengimpor package pandas juga:

```python
import matplotlib.pyplot as plt
import plotnine as p9
import pandas as pd
```

<h3><a name="load-dataset"></a>Load Dataset</h3>

Selanjutnya kita akan membaca data-data yang kita perlukan. Terdapat dua dataset yang akan kita gunakan di modul ini:


1. Data kependudukan DKI Jakarta, yang akan disimpan dalam variabel bernama df_penduduk
2. Data inflasi Indonesia dan Singapura, yang akan disimpan dalam variabel bernama df_inflasi

```python
df_penduduk = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/datakependudukandki-dqlab.csv')
df_inflasi = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/inflasi.csv')
```

<h2><a name="membuat-plot-sederhana"></a>Chapter 2: Membuat Plot Sederhana</h2>

Ide dasar dari grammar of graphics adalah, sebuah grafik statistik merupakan pemetaan dari data ke atribut-atribut aestetik (seperti warna, ukuran, dan bentuk) dari objek-objek geometris (seperti titik dan garis). Pada package plotnine , sebuah grafik terdiri dari komponen-komponen sebagai berikut:

1. <b>data:</b> sebuah data frame yang berisi data yang ingin kita visualisasikan
2. <b>geoms </b>(geometric objects): objek geometris seperti lingkaran, titik, dan teks yang ingin kita lihat di grafik
3. <b>aesthetics:</b> pemetaan dari data ke geometric objects seperti posisi, ukuran, warna, bentuk, dll.


Di <code>plotnine</code> , ketiga komponen tersebut dihubungkan dengan operator + .


Di bagian ini, kita akan membuat plot pertama kita dengan menggunakan data yang ada di variabel df_penduduk .
Kita akan mencoba untuk membuat visualisasi jumlah penduduk di DKI Jakarta untuk masing-masing kabupaten/kota.

```python
df_penduduk.head()
```

<h3><a name="menjalankan-fungsi-ggplot"></a>Menjalankan Fungsi ggplot</h3>

Pertama-tama, kita akan membuat plot terlebih dahulu. Untuk membuat plot, kita akan menggunakan fungsi ggplot . Untuk membuat sebuah plot kita perlu ggplot data apa yang ingin kita visualisasikan, dengan argumen data .


Jalankan fungsi ggplot dan tambahkan argumen data dengan nilai berupa data frame data kependudukan yang telah kita definisikan sebelumnya!

Note : Karena kita belum mendefinisikan komponen-komponen lain seperti aestetik dan objek geometris, maka kita akan mendapatkan sebuah plot kosong seperti berikut ini:

```python
import matplotlib.pyplot as plt
from plotnine import *

ggplot(data=df_penduduk).draw()
plt.show()
```

<h3><a name="menambahkan-variabel"></a>Menambahkan Variabel</h3>

Selanjutnya, kita perlu mendefinisikan variabel-variabel apa yang ingin kita visualisasikan di plot, dengan mendefinisikan aestetik. Aestetik dapat didefinisikan dengan menggunakan fungsi <code>aes()</code>. Untuk x-axis, kita akan menggunakan variabel NAMA KABUPATEN/KOTA , sementara untuk y-axis kita akan gunakan variabel JUMLAH .


Dengan fungsi aes , tambahkan argumen x dengan variabel berupa kolom yang berisi nama kabupaten/kota di dataset penduduk, dan untuk argumen y gunakan variabel berupa kolom berisi jumlah penduduk di dataset penduduk.

```python
import matplotlib.pyplot as plt
from plotnine import *

(ggplot(data=df_penduduk)
+ aes(x='NAMA KABUPATEN/KOTA', y='JUMLAH')
).draw()
plt.show()
```

<h3><a name="mendefinisikan-objek-geometris"></a>Mendefinisikan Objek Geometris</h3>

Kita sudah memiliki plot dengan NAMA KABUPATEN/KOTA sebagai x-axis dan JUMLAH sebagai y-axis. Namun kita belum mendefinisikan objek geometris apa yang ingin kita gunakan, jadi di plot tersebut belum ada data yang ditampilkan.


Ketika menggabungkan dua atau lebih komponen, perlu diperhatikan bahwa:
1. Kita menggunakan operator +
2. Kita perlu menggunakan tanda kurung di keseluruhan fungsi kita, karena kalau tidak, Python akan menganggap itu sebagai error.

Jadi, alih-alih menuliskan:
<br><code>ggplot(data=df_penduduk)</code>
<br><code>+ aes(x='NAMA KABUPATEN/KOTA', y='JUMLAH')</code>


Kita harus menuliskan tanda kurung di sekitar fungsinya:
<br><code>(ggplot(data=df_penduduk)</code>
<br><code>+ aes(x='NAMA KABUPATEN/KOTA', y='JUMLAH'))</code>

Sebagai contoh, kita menggunakan bars untuk membuat visualisasi jumlah penduduk per kabupaten/kota. Kita akan menggunakan geom <code>geom_col()</code> . Tinggi dari bars yang dihasilkan oleh <code>geom_col()</code> merepresentasikan data yang telah kita definisikan di y-axis.

Tambahkan geom untuk membuat bar chart di kode di bawah ini, agar kita dapat memunculkan data jumlah penduduk.

```python
import matplotlib.pyplot as plt
from plotnine import *

(ggplot(data=df_penduduk)
+ aes(x='NAMA KABUPATEN/KOTA', y='JUMLAH')
+ geom_col()
).draw()
plt.show()
```

Notes : Lebih lanjut tentang <code>geom_col()</code> dapat dibaca di <a href=https://plotnine.readthedocs.io/en/stable/generated/plotnine.geoms.geom_col.html>https://plotnine.readthedocs.io/en/stable/generated/plotnine.geoms.geom_col.html</a>

<h3><a name="membuat-horizontal-bar-chart"></a>Membuat Horizontal Bar Chart</h3>

Pada plot sebelumnya kita sudah mengenal komponen-komponen dasar yang digunakan di plotnine , yaitu data, aestetik, dan geoms. Selanjutnya kita akan mempelajari hal-hal yang kita dapat lakukan untuk membuat visualisasi data yang tampak lebih rapi dan menarik.

Satu hal yang mungkin kita perhatikan adalah label di x-axis yang berdempetan. Hal ini karena label-label di x-axis kita mempunyai nama yang cukup panjang. Salah satu solusi yang dapat dilakukan adalah membuat horizontal bar chart. Di plotnine, kita dapat menggunakan <code>coord_flip()</code> agar x-axis di grafik berubah posisi dari horizontal menjadi vertikal, dan y-axis berubah posisi dari vertikal ke horizontal.

Tambahkan fungsi yang diperlukan agar kita bisa mendapatkan horizontal bar chart dari plot yang telah kita buat sebelumnya.

```python
import matplotlib.pyplot as plt
from plotnine import *

(ggplot(data=df_penduduk)
+ aes(x='NAMA KABUPATEN/KOTA', y='JUMLAH')
+ geom_col()
+ coord_flip()
).draw()

plt.tight_layout()
plt.show()
```

Note : Lebih lanjut tentang fungsi tersebut dapat dilihat di <a href="https://plotnine.readthedocs.io/en/stable/generated/plotnine.coords.coord_flip.html?highlight=coord_flip">https://plotnine.readthedocs.io/en/stable/generated/plotnine.coords.coord_flip.html?highlight=coord_flip</a>

<h3><a name="kustomasi-judul-dan-label"></a>Kustomasi Judul dan Label</h3>

Kita dapat menambahkan judul dan mengubah label pada axis di grafik dengan fungsi <code>labs()</code> . Berikut beberapa argumen yang dapat kita gunakan:

<ul>
    <li><b>title</b> : menambahkan judul plot</li>
    <li><b>x</b> : mendefinisikan label pada x-axis</li>
    <li><b>y</b> : mendefinisikan label pada y-axis</li>
</ul>

Gunakan fungsi labs untuk:
<ul>
    <li>Menambahkan judul grafik yaitu "Jumlah penduduk per kabupaten/kota di DKI Jakarta (2013)"</li>
    <li>Mengubah label "JUMLAH" menjadi "Jumlah Penduduk"</li>
    <li>Mengubah label "NAMA KABUPATEN/KOTA" menjadi "Kabupaten/Kota"</li>
</ul>

```python
import matplotlib.pyplot as plt
from plotnine import *
import plotnine

plotnine.options.figure_size=(12, 4.8)
(ggplot(data=df_penduduk)
+ aes(x='NAMA KABUPATEN/KOTA', y='JUMLAH')
+ geom_col()
+ coord_flip()
+ labs(title='Jumlah penduduk per kabupaten/kota di DKI Jakarta (2013)',
x='Kabupaten/Kota',
y='Jumlah Penduduk')
).draw()
plt.show()
```

<h3><a name="menampilkan-warna-berbeda"></a>Menampilkan Warna Berbeda</h3>

Bagaimana jika kita ingin menampilkan bar chart yang memiliki warna berbeda untuk perempuan dan laki-laki? Kita dapat membuatnya dengan mendefinisikan argumen <code>fill</code> di fungsi <code>aes</code> .

Pada argumen <code>fill</code> di fungsi <code>aes</code>, berikan nilai berupa nama kolom jenis kelamin di data kependudukan kita.

```python
import matplotlib.pyplot as plt
from plotnine import *
import plotnine

plotnine.options.figure_size=(10, 3.6)
(ggplot(data=df_penduduk)
+ aes(x='NAMA KABUPATEN/KOTA', y='JUMLAH', fill='JENIS KELAMIN')
+ geom_col()
+ coord_flip()
+ labs(title='Jumlah penduduk per kabupaten/kota di DKI Jakarta (2013)',
       x='Kabupaten/Kota',
       y='Jumlah Penduduk')
).draw()
plt.show()
```

<h3><a name="membuat-grafik-dengan-variabel-berbeda"></a>Membuat Grafik dengan Variabel Berbeda</h3>

Sekarang kita akan berlatih untuk membuat grafik serupa namun dengan variabel yang berbeda.

Buatlah grafik horizontal bar chart dengan ketentuan sebagai berikut:

<ul>
    <li>Bar chart menggambarkan jumlah penduduk per kelurahan di kecamatan Cengkareng. </li>
    <li>Judul grafik yaitu "Jumlah penduduk per kelurahan di Kecamatan Cengkareng (2013)"</li>
    <li>Pada axis yang berisi nama kelurahan, berikan label "Kelurahan"</li>
    <li>Pada axis yang berisi jumlah penduduk, berikan label "Jumlah Penduduk"</li>
    <li>Berikan warna yang berbeda untuk jenis kelamin laki-laki dan perempuan.</li>
</ul>

```python
import matplotlib.pyplot as plt
from plotnine import *
import plotnine

plotnine.options.figure_size=(10, 3.6)
(ggplot(data=df_penduduk[df_penduduk['NAMA KECAMATAN'] == 'CENGKARENG'])
+ aes(x='NAMA KELURAHAN', y='JUMLAH', fill='JENIS KELAMIN')
+ geom_col()
+ coord_flip()
+ labs(title='Jumlah penduduk per kelurahan di DKI Jakarta (2013)',
      x='Kelurahan',
      y='Jumlah Penduduk')
).draw()
plt.show()
```

<h3><a name="memisahkan-grafik"></a>Memisahkan Grafik</h3>

Contoh plot yang telah kita buat adalah bar chart yang berbentuk stacked, dalam artian untuk data jenis kelamin laki-laki dan perempuan, keduanya bertumpuk di satu bar. Bagaimana kalau kita ingin memisahkannya dan meletakannya di samping satu sama lain?

Kita dapat menggunakan argumen position di fungsi <code>geom_col()</code>. Untuk memisahkan grafik tersebut kita dapat mendefinisikan <code>position='position_dodge'</code>.

```python
import matplotlib.pyplot as plt
from plotnine import *
import plotnine

plotnine.options.figure_size=(10, 3.6)
(ggplot(data=df_penduduk[df_penduduk['NAMA KECAMATAN'] == 'CENGKARENG'])
+ aes(x='NAMA KELURAHAN', y='JUMLAH', fill='JENIS KELAMIN')
+ geom_col(position='position_dodge')
+ coord_flip()
+ labs(title='Jumlah penduduk per kelurahan di DKI Jakarta (2013)',
x='Kelurahan',
y='Jumlah Penduduk')
).draw()
plt.show()
```

<h2><a name="mengenal-berbagai-geom"></a>Chapter 3: Mengenal Berbagai Geom</h2>

<h3><a name="membuat-scatterplot"></a>Membuat Scatterplot</h3>

Scatterplot biasanya digunakan untuk melihat hubungan antara dua variabel. Kita dapat menggunakan scatterplot untuk melihat hubungan antara jumlah penduduk di suatu kelurahan dengan luas kelurahan tersebut. Untuk mempermudah, kita akan mendefinisikan data frame baru, <b>df_penduduk_luas_jumlah</b> , sebagai rangkuman dari kolom nama kelurahan, luas wilayah, dan jumlah penduduk.

```python
df_penduduk_luas_jumlah = df_penduduk.groupby(['NAMA KELURAHAN', 'LUAS WILAYAH (KM2)'])[['JUMLAH']].agg('sum').reset_index()
df_penduduk_luas_jumlah
```

Untuk membuat scatterplot, kita dapat menggunakan <code>geom_point()</code> .
<ul>
    <li>Buatlah sebuah scatterplot dengan menggunakan data di data frame df_penduduk_luas_jumlah.</li>
    <li>Gunakan <code>ggplot()</code> untuk mendefinisikan data, <code>aes()</code> untuk menentukan aestetik, dan <code>geom_point()</code> sebagai geom -nya.</li>
    <li>Variabel di x-axis adalah jumlah penduduk, sementara di y-axis adalah luas wilayah (km persegi).</li>
</ul>

```python
import matplotlib.pyplot as plt
from plotnine import *

(ggplot(data=df_penduduk_luas_jumlah)
+ aes(x='JUMLAH', y='LUAS WILAYAH (KM2)')
+ geom_point()
).draw()
plt.show()
```

Kita juga dapat memberikan warna untuk menggambarkan besarnya nilai dari variabel yang telah kita tentukan.

Pada fungsi <code>aes()</code>, gunakan argumen color agar kita dapat memberikan warna yang merepresentasikan besarnya nilai dari variabel jumlah penduduk.

```python
import matplotlib.pyplot as plt
from plotnine import *
import plotnine

plotnine.options.figure_size=(10, 3.6)
(ggplot(data=df_penduduk_luas_jumlah)
+ aes(y='LUAS WILAYAH (KM2)', x='JUMLAH', color='JUMLAH')
+ geom_point()
).draw()
plt.show()
```

<h3><a name="membuat-histogram"></a>Membuat Histogram</h3>

Histogram berguna untuk menggambarkan distribusi data atau merangkum data. Di sini, kita akan menggunakan histogram untuk melihat distribusi luas wilayah di data kita. Untuk membuat histogram di plotnine , kita dapat menggunakan <code>geom_histogram()</code>.

Buatlah sebuah histogram menggunakan data df_penduduk yang menggambarkan distribusi luas wilayah di dataset kependudukan! x-axis di histogram ini berupa luas wilayah (dalam kilometer persegi), sementara y-axis-nya berupa jumlah kelurahan yang memiliki luas wilayah tersebut.

```python
import matplotlib.pyplot as plt
from plotnine import *

(ggplot(data=df_penduduk)
+ aes(x='LUAS WILAYAH (KM2)')
+ geom_histogram()
).draw()
plt.show()
```

Terkadang, dalam pembuatan histogram, kita ingin tinggi bar di plot kita memiliki nilai maksimal 1. Oleh karena itu kita perlu melakukan kalkulasi. Kita dapat melakukan kalkulasi tersebut langsung di aestetik dengan menggunakan bantuan <b>stat</b> (statistical transformation atau transformasi statistik).

Secara default, ketika kita menggunakan histogram sebenarnya plotnine telah mendefinisikan <code>y='stat(count)'</code> yang berarti y akan merepresentasikan count dari luas wilayah tersebut. Karena sekarang kita ingin supaya tinggi bar kita memiliki nilai maksimal 1, kita perlu mendefinisikan y secara eksplisit dengan definisi kita sendiri.

Definisikan y di fungsi <code>aes()</code> dan masukkan <code>'stat(count/max(count))'</code> sebagai nilainya.

```python
import matplotlib.pyplot as plt
from plotnine import *

(ggplot(data=df_penduduk)
+ aes(x='LUAS WILAYAH (KM2)', y='stat(count/max(count))')
+ geom_histogram()
).draw()
plt.show()
```

<h3><a name="membuat-boxplot"></a>Membuat Boxplot</h3>

Boxplot biasanya digunakan untuk melihat seberapa spread out dataset yang kita miliki. plotnine telah menyediakan geom bernama <code>geom_boxplot()</code>.

Buatlah sebuah boxplot yang menggambarkan persebaran dari data jumlah penduduk di masing-masing
kabupaten/kota!

```python
import matplotlib.pyplot as plt
from plotnine import *

(ggplot(data=df_penduduk)
+ aes(x='NAMA KABUPATEN/KOTA', y='JUMLAH')
+ geom_boxplot()
+ coord_flip()
).draw()
plt.tight_layout()
plt.show()
```

<h3><a name="membuat-line-chart"></a>Membuat Line Chart</h3>

Line chart biasanya digunakan untuk memvisualisasikan pergerakan data seiring berjalannya waktu. Dalam pembuatan line chart kali ini, kita akan menggunakan data inflasi. geom yang dapat kita gunakan adalah <code>geom_line()</code> . Dengan line chart yang kita buat, kita akan mengetahui pergerakan inflasi (dalam persen) di negara Indonesia dan Singapura dari bulan ke bulan, mulai dari Januari sampai Oktober 2017.

Namun sebelum itu kita perlu mengubah tipe data kolom Bulan di <b>df_inflasi</b> dari string menjadi <i>datetime</i>. Hal ini agar plotnine dapat mengetahui bahwa Bulan adalah kolom yang menggambarkan waktu. Sehingga, plotnine akan secara otomatis mengurutkan inflasi di data kita berdasarkan urutan bulan.

```python
df_inflasi['Bulan'] = pd.to_datetime(df_inflasi['Bulan'])
df_inflasi
```

Sebagai awal, kita akan membuat line chart untuk negara Indonesia terlebih dahulu. Buatlah line chart yang menunjukkan pergerakan inflasi (dalam persen) dari bulan ke bulan untuk negara Indonesia dengan menggunakan <code>geom_line()</code>.

Gunakan <code>aes()</code> untuk mendefinisikan variabel di x-axis dan y-axis. Di x-axis variabelnya adalah Bulan, serta di y-axis variabelnya adalah Inflasi.

```python
import matplotlib.pyplot as plt
from plotnine import *

(ggplot(data=df_inflasi[df_inflasi['Negara']=='Indonesia'])
+ aes(x='Bulan', y='Inflasi')
+ geom_line()
).draw()
plt.show()
```

Bagaimana jika kita ingin membuat line chart untuk negara Singapura dan Indonesia di plot yang sama? Kita dapat mendefinisikan argumen color di fungsi <code>aes()</code>.

Isi argumen color dengan kolom negara yang ada di df_inflasi , agar kita mempunyai dua buah garis di plot kita: garis untuk Indonesia dan Singapura.

```python
import matplotlib.pyplot as plt
from plotnine import *
import plotnine

plotnine.options.figure_size=(10, 3.6)
(ggplot(data=df_inflasi)
+ aes(x='Bulan', y='Inflasi', color='Negara')
+ geom_line()
).draw()
plt.show()
```

Kita bisa lihat bahwa label di x-axis kita bertabrakan. Salah satu solusi yang dapat dilakukan adalah mengubah ukuran dari plot yang kita buat.
Gunakan fungsi <code>theme()</code> dari plotnine dan berikan argumen <code>figure_size</code> untuk mendefinisikan grafik berukuran 10x5. figure_size menerima nilai berupa tuple yang berisi (panjang, tinggi) dalam inch.

```python
import matplotlib.pyplot as plt
from plotnine import *

(ggplot(data=df_inflasi)
+ aes(x='Bulan', y='Inflasi', color='Negara')
+ geom_line()
+ theme(figure_size=(10, 5))
).draw()
plt.show()
```

Note : Lebih lanjut tentang <code>theme()</code> dapat dibaca di <a href=https://plotnine.readthedocs.io/en/stable/generated/plotnine.themes.theme.html>https://plotnine.readthedocs.io/en/stable/generated/plotnine.themes.theme.html</a>

```python
# membuat line chart dengan mapping aes
import matplotlib.pyplot as plt
import plotnine as p9

(p9.ggplot(data=df_inflasi[df_inflasi['Negara']=='Indonesia'], 
           mapping=p9.aes(x='Bulan', y='Inflasi'))
+ p9.geom_line()
).draw()
plt.show()
```
